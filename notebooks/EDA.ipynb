```python
data['Email Content']=data['Email Content'].str.lower #convert all the text to lower text
```

```python
def whitespace(sentence):
     new  =re.sub(r'\s+', ' ', sentence).strip()
     return new

data['Email Content']=data['Email Content'].apply(lambda x: whitespace(x)) # normalizess the whitespaces
```
